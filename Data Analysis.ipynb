```python
#the explained variance ratio by each component tells us how much information each component retains independently, 
#while the total variance explained by all components gives us an overall measure of how much information is retained 
#collectively by the selected principal components. 
# Feature selection using PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assuming 'df' is your DataFrame and 'features' are the features you want to include in PCA
features = ['Ratings_out_of_5_Restaurant',
       'Price_Restaurant', 
       'Distance_to_Place_0_from__Restaurant','hotel_star_rating_Hotel',
       'mmt_review_score_Hotel', 'Price_Hotel', 'Distance_to_Place_form_Hotel', 'budget',
       'budget_level']  # Add more features as needed

# Extract features from DataFrame
X = df[features]

# Standardize the features by scaling them to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Specify the number of components to keep (can be a fraction of total features or a specific number)
n_components = 9

# Perform PCA
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame with principal components
principal_components = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, n_components + 1)])

# Concatenate principal components with original DataFrame if needed
df_with_pca = pd.concat([df, principal_components], axis=1)

# Variance explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained variance ratio by each component:", explained_variance_ratio)

# Total variance explained by all principal components
total_variance_explained = sum(explained_variance_ratio)
print("Total variance explained by all components:", total_variance_explained)
```

```python
import seaborn as sns


# Assuming 'df' is your DataFrame containing the numerical features
numerical_features = df[['Ratings_out_of_5_Restaurant',
                         'Price_Restaurant', 
                         'Distance_to_Place_0_from__Restaurant',
                         'hotel_star_rating_Hotel',
                         'mmt_review_score_Hotel', 
                         'Price_Hotel', 
                         'Distance_to_Place_form_Hotel', 
                         'budget',
                         'budget_level']]

# Calculate the correlation matrix between numerical features
correlation_matrix = numerical_features.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add title and labels
plt.title('Correlation Heatmap of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Features')

# Display the heatmap
plt.show()
```
