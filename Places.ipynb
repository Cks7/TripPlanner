```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Assuming you have three DataFrames: places_data, restaurants_data, and hotels_data

# Drop unnecessary columns for the matching process
places_features = places_data[['Longitude', 'Latitude']]
restaurants_features = restaurants_data[['Longitude', 'Latitude']]
hotels_features = hotels_data[['Longitude', 'Latitude']]

# Use Nearest Neighbors to find the nearest place for each restaurant/hotel
knn = NearestNeighbors(n_neighbors=3, metric='haversine')
knn.fit(places_features)

# Find the nearest place for each restaurant
distances_restaurants, indices_restaurants = knn.kneighbors(restaurants_features)

# Find the nearest place for each hotel
distances_hotels, indices_hotels = knn.kneighbors(hotels_features)

# Create a new column 'Nearest_Place_Index' to store the index of the nearest place
restaurants_data['Nearest_Place_Index'] = indices_restaurants.flatten()
hotels_data['Nearest_Place_Index'] = indices_hotels.flatten()

# Reset index before merging to avoid length mismatch issue
restaurants_data = restaurants_data.reset_index(drop=True)
hotels_data = hotels_data.reset_index(drop=True)

# Merge the 'places_data' DataFrame with 'restaurants_data' and 'hotels_data' based on the 'Nearest_Place_Index'
restaurants_data = pd.merge(restaurants_data, places_data, left_on='Nearest_Place_Index', right_index=True, suffixes=('_restaurant', '_place'))
hotels_data = pd.merge(hotels_data, places_data, left_on='Nearest_Place_Index', right_index=True, suffixes=('_hotel', '_place'))

# Now, 'restaurants_data' and 'hotels_data' should have additional columns from the nearest place in 'places_data'.
```

```python
# Merge 'restaurants_data' and 'hotels_data' based on the 'Nearest_Place_Index'
merged_data = pd.merge(restaurants_data, hotels_data, left_on='Nearest_Place_Index', right_on='Nearest_Place_Index', suffixes=('_restaurant', '_hotel'))

# Merge the combined data with 'places_data' based on the 'Nearest_Place_Index'
final_data = pd.merge(merged_data, places_data, left_on='Nearest_Place_Index', right_index=True, suffixes=('_combined', '_place'))

# Drop unnecessary columns
final_data = final_data.drop(columns=['Nearest_Place_Index'])

# Save the final dataset to a new CSV file
final_data.to_csv('final_dataset.csv', index=False)
```
