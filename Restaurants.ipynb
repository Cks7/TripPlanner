```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors

# Assuming you have loaded your restaurant dataset into a DataFrame called 're_data'
# For simplicity, we'll only consider a few relevant features.

# Selecting relevant features
restaurant_data = re_data[['Ratings_out_of_5', 'price', 'Price Level']]

# Convert categorical variables (Locality) to numerical using one-hot encoding
restaurant_data = pd.get_dummies(re_data['Locality']).join(restaurant_data)

# Assuming 'user_preferences' is a dictionary containing user preferences
user_preferences = {
    'Ratings_out_of_5': 4.5,
    'Locality_Bhugaon': 1,  # Example locality preference (modify based on your data)
    'Price Level': 2
}

# Creating a DataFrame for user preferences
user_df = pd.DataFrame([user_preferences])

# Extracting categorical columns from the user_df
categorical_columns = restaurant_data.select_dtypes(include=['category']).columns

# Ensure that the categorical columns in user_df have the same categories as in restaurant_data
for col in categorical_columns:
    user_df[col] = user_df[col].astype('category')

    # Set the categories explicitly to match the restaurant_data
    user_df[col] = user_df[col].cat.set_categories(restaurant_data[col].cat.categories)

# Concatenate user preferences with the restaurant data
combined_data = pd.concat([user_df, restaurant_data], axis=1)

# Fill missing values with 0 (for areas where the user hasn't provided a preference)
combined_data = combined_data.fillna(0)

# Check data types of columns in combined_data
print("Data Types of Columns in combined_data:")
print(combined_data.dtypes)

# Calculating cosine similarity for numeric columns
numeric_columns = combined_data.select_dtypes(include=['float64', 'int64']).columns
cosine_sim = cosine_similarity(combined_data[numeric_columns])

# Using Nearest Neighbors to find similar restaurants
knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(cosine_sim)

# Recommending restaurants for the user
user_index = 0  # Assuming the user is the first row in the combined_data DataFrame
distances, indices = knn.kneighbors(combined_data.iloc[user_index][numeric_columns].values.reshape(1, -1))

# Displaying recommended restaurants
recommended_restaurants = restaurant_data.iloc[indices[0][1:]]
print("Recommended Restaurants:")
print(recommended_restaurants[['Ratings_out_of_5', 'price', 'Price Level']])
```
