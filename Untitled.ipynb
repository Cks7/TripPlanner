```python
import scrapy

class HotelPriceSpider(scrapy.Spider):
    name = "hotel_price_spider"
    allowed_domains = ["www.makemytrip.com"]

    def __init__(self, hotel_names):
        self.hotel_names = hotel_names

    def start_requests(self):
        for hotel_name in self.hotel_names:
            url = f"https://www.makemytrip.com/hotels/{hotel_name}"
            yield scrapy.Request(url, callback=self.parse)

    def parse(self, response):
        hotel_name = response.css(".hotel_name_header span.hotel_name").xpath("text()").get()
        price = response.css(".hotel_price_box .price").xpath("text()").get()

        yield {
            "hotel_name": hotel_name,
            "price": price,
        }
```

```python
import requests
from bs4 import BeautifulSoup
import csv

urls = ["https://www.booking.com/hotels/in/pune"]

data = []

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    hotel_name = soup.find("h2", class_="bui-hotel-name").text
    price = soup.find("span", class_="bui-price-display__value").text
    rating = soup.find("span", class_="bui-rating-value__text").text

    data.append([hotel_name, price, rating])

with open("hotels3.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Hotel Name", "Price", "Rating"])
    writer.writerows(data)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import haversine_distances
from math import radians

# Load your dataset
df = pd.read_csv('final_dataset.csv')

# Example user input
user_budget = 1000

# Filter based on both restaurant budget and hotel price
filtered_data = df[(df['price'] <= user_budget) & (df['Price_x'] <= user_budget)]

# Select the first restaurant based on budget
selected_restaurant = filtered_data.iloc[0]

# Calculate Haversine distance between the selected restaurant and hotels/places
def calculate_distance(row):
    lat1, lon1 = selected_restaurant['Latitude_restaurant'], selected_restaurant['Longitude_restaurant']
    lat2, lon2 = row['Latitude_hotel'], row['Longitude_hotel']
    dist = haversine_distances([[radians(lat1), radians(lon1)], [radians(lat2), radians(lon2)]]).item(1) * 6371000  # Earth radius in meters
    return dist

# Apply the distance calculation to hotels
filtered_data['distance_to_restaurant'] = filtered_data.apply(calculate_distance, axis=1)

# Rank hotels based on distance
ranked_hotels = filtered_data[filtered_data['Name_hotel'].notnull()].sort_values(by='distance_to_restaurant')

# Select the nearest hotel
nearest_hotel = ranked_hotels.iloc[0]

# Apply the distance calculation to places
filtered_data['distance_to_restaurant'] = filtered_data.apply(calculate_distance, axis=1)

# Rank places based on distance
ranked_places = filtered_data[filtered_data['Name'].notnull()].sort_values(by='distance_to_restaurant')

# Select the nearest place
nearest_place = ranked_places.iloc[0]

print("Recommended Restaurant:", selected_restaurant['Restaurant_Name'])
print("Nearest Hotel:", nearest_hotel['property_name'])
print("Nearest Place:", nearest_place['Name'])
```
